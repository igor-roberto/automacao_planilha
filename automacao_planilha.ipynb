```python
import pandas as pd
import openpyxl

from selenium import webdriver


planilha = pd.read_excel("produtos_ficticios.xlsx")

nav= webdriver.Chrome()


for i, nome_produto in enumerate(planilha["Nome do produto"]):

    descricao = planilha.loc[i,"Descrição"]

    categoria = planilha.loc[i,"Categoria"]

    codigo_produto = planilha.loc[i,"Código do produto"]

    peso = planilha.loc[i,"Peso (em kg)"]

    dimensoes = planilha.loc[i,"Dimensões (L x A x P)"]

    preco = planilha.loc[i,"Preço"]

    qtd_estoque = planilha.loc[i,"Quantidade em estoque"]

    data_val = planilha.loc[i,"Data de validade"]

    cor = planilha.loc[i,"Cor"]

    tamanho = planilha.loc[i,"Tamanho"]

    material = planilha.loc[i,"Material"]

    fabricante = planilha.loc[i,"Fabricante"]

    pais_origem = planilha.loc[i,"País de origem"]

    observacoes = planilha.loc[i,"Observações"]

    codigo_barras = planilha.loc[i,"Código de barras"]

    loc_armazem = planilha.loc[i,"Localização no armazém"]
    

    nav.get("https://forms.gle/swdQo6HoWrBr9pW26")
    

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[1]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(nome_produto)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[2]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(descricao)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[3]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(categoria)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[4]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(codigo_produto))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[5]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(peso))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[6]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(dimensoes))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[7]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(preco))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[8]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(qtd_estoque))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[9]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(data_val)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[10]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(cor)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[11]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(tamanho)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[12]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(material)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[13]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(fabricante)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[14]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(pais_origem)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[15]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(observacoes)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[16]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(str(codigo_barras))

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[2]/div[17]/div/div/div[2]/div/div[1]/div/div[1]/input').send_keys(loc_armazem)

    nav.find_element("xpath",'//*[@id="mG61Hd"]/div[2]/div/div[3]/div[1]/div[1]/div/span/span').click()
```
